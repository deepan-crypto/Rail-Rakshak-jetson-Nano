```python
import os
import yaml

# 1. Grab the exact path to the Roboflow data.yaml file
yaml_path = os.path.join(dataset.location, "data.yaml")
print(f"âœ… Ready to train using config: {yaml_path}")

# --- YAML Correction Start ---
# Load the current data.yaml
with open(yaml_path, 'r') as f:
    data_yaml_content = yaml.safe_load(f)

print("Original data.yaml content:")
print(yaml.dump(data_yaml_content, default_flow_style=False))

# Update the 'path' to the dataset root, which is dataset.location
# dataset.location was set by Roboflow download to /content/Railway-Track-Fault-detection-4
data_yaml_content['path'] = dataset.location

# Ensure 'train', 'val', 'test' are relative paths to the 'path'
# Roboflow usually generates paths like 'train/images', 'valid/images'
# but sometimes they can be absolute or include the parent directory name.
# We will explicitly set them to 'train/images', 'valid/images', etc.
# assuming the Roboflow YOLOv5 dataset structure.
if 'train' in data_yaml_content:
    data_yaml_content['train'] = 'train/images'
if 'val' in data_yaml_content:
    data_yaml_content['val'] = 'valid/images'
if 'test' in data_yaml_content: # Test set might not always be present
    data_yaml_content['test'] = 'test/images'

# Save the updated data.yaml
with open(yaml_path, 'w') as f:
    yaml.dump(data_yaml_content, f, default_flow_style=False)

print("\nUpdated data.yaml content:")
with open(yaml_path, 'r') as f:
    print(f.read())
# --- YAML Correction End ---

# 2. Make sure we are inside the yolov5 folder
%cd /content/yolov5

# 3. Disable Weights & Biases so it doesn't freeze asking for a login again
os.environ["WANDB_DISABLED"] = "true"

# 4. Start the training!
# We are using yolov5s (Small) because it is the best fit for Jetson Nano's memory
print("ğŸš€ Starting YOLOv5 Training Phase...")
!python train.py --img 640 --batch 16 --epochs 50 --data {yaml_path} --weights yolov5s.pt --name rail_rakshak
```
